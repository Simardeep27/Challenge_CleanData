```python
def converter_null(df,Column):
    Column_val=df[['{}'.format(Column)]]
    ordinal_encoder=OrdinalEncoder()
    encoded=ordinal_encoder.fit_transform(Column_val)
    df['{}'.format(Column)]=encoded
    
def replace_nan(df,column):
    df_copy=df.copy()
    df_predict=df_copy[['Business Title','Civil Service Title','Division/Work Unit','Job Description','Minimum Qual Requirements','Hours/Shift','Residency Requirement','AIFuture','Age','loan_status','dti','last_pymnt_d','last_pymnt_amnt','{}'.format(column)]]
    train_test_vals=df_predict[df_predict['{}'.format(column)].notna()]
    null_vals=df_predict[df_predict.isna().any(axis=1)]
    df_predict.head()
    converter_null(train_test_vals,column)
    X_set=train_test_vals[['Business Title','Civil Service Title','Division/Work Unit','Job Description','Minimum Qual Requirements','Hours/Shift','Residency Requirement','AIFuture','Age','loan_status','dti','last_pymnt_d','last_pymnt_amnt']]
    Y_set=train_test_vals[['{}'.format(column)]]
    X_train,X_test,y_train,y_test=train_test_split(X_set,Y_set)
    y_train=np.reshape(y_train,newshape=(y_train.shape[0],))
#     gsc=GridSearchCV(estimator=RandomForestClassifier(),
#                     param_grid={
#                         'n_estimators': [10,50,100,150,200,250,300],
#                         'max_depth':[2,4,6,8,10,12],
#                         'min_samples_split':[2,4,6,8,10],
#                         'min_samples_leaf':[2,4,6,8,10]
#                     },cv=5,scoring='accuracy')
#     grid_result=gsc.fit(X_train,y_train)
#     print(grid_result)
#     best_params = grid_result.best_params_
#     print(best_params)
    DC=RandomForestClassifier().fit(X_train,y_train)
    y_pred=DC.predict(X_test)
    print(acc(y_test,y_pred))
    predict_feat=null_vals[['Business Title','Civil Service Title','Division/Work Unit','Job Description','Minimum Qual Requirements','Hours/Shift','Residency Requirement','AIFuture','Age','loan_status','dti','last_pymnt_d','last_pymnt_amnt']]
    predicted_val=DC.predict(predict_feat)
    null_vals['{}'.format(column)]=predicted_val
    df_predict=train_test_vals.append(null_vals).sort_index()
    df_copy['{}'.format(column)]=df_predict['{}'.format(column)]
    return df_copy
```

```python
def feature_check(df,feature):
    df_copy=df[df['{}'.format(feature)].notna()]
    converter_null(df_copy,'{}'.format(feature))
    df_vals=df_copy.corr()
    df_val=df_vals.index.where(np.abs(df_vals['{}'.format(feature)])>0.03)
    df_val=df_val[df_val.notna()]
    df_val=df_val.drop('{}'.format(feature))
    for i in df_val:
        print(i)
    return df_vals
```
